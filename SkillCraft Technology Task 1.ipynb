```python
# Filter one country (example: India)
india = df[df["Country"] == "India"]
```

```python
# Age groups & values
age_groups = ["0-4 years", "5-14 years", "15-24 years", "25-64 years", "65+"]
values = india[age_groups].values[0]
```

```python
# Create bar chart
bars = plt.bar(age_groups, values,color='skyblue')
plt.title("Distribution of Population by Age Group",fontsize=12,fontweight='bold',color='green')
plt.xlabel("Age Group",fontweight='bold')

```
